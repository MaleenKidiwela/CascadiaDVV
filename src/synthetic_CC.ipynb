## Theory 
We will compute synthetic Cross-correlations between vertical components, using the formulation given in Tomasetto et al.(2024). Using the representation theorem and assuming spatially uncorrelated sources one can write the cross-correlation function between sensors A and B as:

$$ C(r_A, r_B, t) = \mathcal{FT}^{-1}\left[ \int_{\partial D} G(r_A, r, f) G^*(r_B, r, f) S(r,f) dr \right] $$

with:
- $G(r_A, r, f)$ the Green's function between a source in $ r $ and station A in $ r_A $. Here computed in AxiSEM.
- $S(r,f)$ the source PSD at position $ r $. Here given by the square of equivalent vertical force amplitude computed in microseismic_sources.ipynb.
- $^*$ denotes the complex conjugate.
- $\partial D$ the domain of potential sources, here, the ocean's surface.
- $\mathcal{FT}^{-1}$ the Fourier transform.

#### References 

- [Nissen-Meyer, T., van Driel, M., Stähler, S. C., Hosseini, K., Hempel, S., Auer, L., ... & Fournier, A. (2014). AxiSEM: broadband 3-D seismic wavefields in axisymmetric media. Solid Earth, 5(1), 425-445.](https://se.copernicus.org/articles/5/425/2014/)

- [Tomasetto, L., Boué, P., Stehly, L., Ardhuin, F., & Nataf, H. C. (2024). On the stability of mantle‐sensitive P‐wave interference during a secondary microseismic event. Geophysical Research Letters, 51(8), e2023GL108018.](https://agupubs.onlinelibrary.wiley.com/doi/10.1029/2023GL108018)

## Synthetic Cross-correlation Computation

```python
## Compute CCF and save it
coords_sta = [lon_sta, lat_sta]
#### Keep Correlations
corr_3h = np.zeros((len(dates_vector), N_time))

### Date Info
for i, date_vect in enumerate(dates_vector):
    ### CCF
    filename_ccf = '/data/wsd02/maleen_data/ww3/CCF/ccf_%s_%d%02d%02d%02d.nc'%(id_sta,  date_vect[0],date_vect[1],date_vect[2],date_vect[3]) # File Name Understandable

    if os.path.exists(filename_ccf):  # Check if file already exists
        print('%s already computed'%filename_ccf)
        corr = xr.open_dataarray(filename_ccf)
    else:  # if not, compute CCF
        t1 = ttime.perf_counter()
        corr, time_corr = ccf_computation_autocorr(coords_sta, path_model, date_vect, spectrum_synth_R, spectrum_synth_Z,extent=extent, fe=fe, N=N_time)
        t2 = ttime.perf_counter()
        print(" Time for station pair: ", id_sta, id_sta, " elapsed: ", t2-t1)
        try:
            corr.to_netcdf('/data/wsd02/maleen_data/ww3/CCF/ccf_%s_%d%02d%02d%02d.nc'%(id_sta, date_vect[0],date_vect[1],date_vect[2],date_vect[3]))
        except:
            print("make directory")
            os.mkdir('/data/wsd02/maleen_data/ww3/CCF')
            corr.to_netcdf('/data/wsd02/maleen_data/ww3/CCF/ccf_%s_%d%02d%02d%02d.nc'%(id_sta, date_vect[0],date_vect[1],date_vect[2],date_vect[3]))
    corr_3h[i, :] = corr.values
    # TAKES 25 MIN #
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the window size (30 days) and the step size (1 day)
window_size = 30
step_size = 1

# Initialize an empty list to store the moving average stacks
stacked_corr = []

# Perform the moving average stacking
for i in range(0, len(corr_3h_normalized) - window_size + 1, step_size):
    window = corr_3h_normalized[i:i + window_size, :]
    stack = np.nanmean(window, axis=0)
    stacked_corr.append(stack)

# Convert the list of stacked data to a numpy array
stacked_corr = np.array(stacked_corr)

# Create a new dates vector corresponding to the stacked data
stacked_dates = dates_corr[:len(stacked_corr)]

# Plot the stacked CCF
fig, ax = plt.subplots(figsize=(15, 10))

# Plot using pcolormesh with the stacked data
im = ax.pcolormesh(corr.time, np.arange(len(stacked_corr)), stacked_corr, cmap='seismic', vmin=-1, vmax=1)

# Add colorbar
plt.colorbar(im, ax=ax)

# Set labels and title
ax.set_xlabel('Lag Time (s)')
ax.set_ylabel('Date')
ax.set_title(f'{id_sta}-{id_sta} CCF Moving Average Stack')

# Set x and y limits
ax.set_xlim(-20, 20)
ax.set_ylim(0, len(stacked_corr))

# Set y-ticks to show fewer date labels for better readability
ax.set_yticks(np.arange(len(stacked_corr))[::50])  # Adjust step size as needed
ax.set_yticklabels(stacked_dates[::50].strftime('%Y-%m-%dT%H'))  # Format date labels

# Save the figure
plt.savefig(f'ccf_{id_sta}_{id_sta}_synth_stack.png', dpi=300, bbox_inches='tight')

plt.show()
```
