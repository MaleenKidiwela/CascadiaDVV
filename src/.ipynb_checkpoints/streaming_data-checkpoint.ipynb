```python
import sys
sys.path.append("./pnwstore/")
from pnwstore.mseed import WaveformClient
from obspy.core.utcdatetime import UTCDateTime
from pnwstore.catalog import PickClient
import torch
import seisbench
from seisbench.models import EQTransformer, PhaseNet
from obspy.core.utcdatetime import UTCDateTime
```

```python
waveformclient = WaveformClient()
pickclient = PickClient("cascadia", "pnwstore")
```

```python
pickclient.query(network = "OO", station = "HYS*", 
             mintime = UTCDateTime("2015-12-11"), 
             maxtime = UTCDateTime("2015-12-13"))
```

```python
model = EQTransformer().from_pretrained("stead")
model.to(torch.device("cuda"))
```

```python
s = waveformclient.get_waveforms(network = "OO", 
                      station = "HYS12", 
                      channel = "EH?", 
                      year = "2015", 
                      doy = "346")
s2 = s.copy().trim(starttime = UTCDateTime("2015-12-12T02:20:00"), endtime = UTCDateTime("2015-12-12T02:30:00"))
s2.filter("bandpass", freqmin = 3, freqmax = 12)
s2.plot()
```

```python
s_prediction = model.annotate(s2)
s_prediction.plot()
```
